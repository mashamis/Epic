```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
a link to a csv file: "https://drive.google.com/file/d/1VB6wqBUa8GixJGRmyjHAhqrrJqH2oI42/view?usp=drive_link"
name: "stock_info.csv"
```

```python
data_first = pd.read_csv('stock_info.csv')
data_first
```

```python
#select only needed columns from the original dataset to form a new one to work with
columns = ['Date', 'Close']
data = data_first[columns]
data
```

```python
#Plot the daily closing price of your chosen stock.
fig = plt.figure(figsize=(25, 10))
plt.scatter(data['Date'], data['Close'], color = 'purple')
plt.plot(data['Date'], data['Close'], '-', color = 'purple')
plt.xlabel('Date')
plt.ylabel('Closing price')

# Set the tick locations and labels on the x-axis
plt.xticks([], [])
plt.show()
```

#Determine if it is a random walk or not. If it is not a random walk, explain why

As we can see from the built plot there is equal chance of going up or down by a random number.
It is also seen that there are both a long-term trend and abrupt changes in the trend.

```python
#Keep the last 5 days of data as a test set. The rest will be the train set.
test = data[-5:]
train = data[:-5]
```

```python
#Forecast the last 5 days using naive forecasting methods, and measure the MSE. Which method is the best?
#Last known value:
last = np.repeat(train['Close'].iloc[-1], len(test))
mse_last = mean_squared_error(test['Close'], last)
print('Last known value error:', mse_last) 

#Naive seasonal:
naive_seasonal = train['Close'][-5:].values
mse_naive_seasonal = mean_squared_error(test['Close'], naive_seasonal)
print("Naive seasonal error: ", mse_naive_seasonal) 

#Historical mean
historical_mean = np.repeat(train['Close'].mean(), len(test))
mse_hist_mean = mean_squared_error(test['Close'], historical_mean)
print("Average forecast error: ", mse_hist_mean) 
```

The best appeared to be the 'Last known value' forecast. But as we can ee from the results all the naive forecasts give pretty bad results, especially the last one, which makes it quite obvious that naive forecasting methods are not very suitable for working with random walks.

```python
#Plot your forecasts
fig = plt.figure(figsize=(20, 10))
plt.plot(train['Date'], train['Close'], label='Train Data')
plt.plot(test['Date'], test['Close'], label='Test Data')
plt.plot(test['Date'], last, label='Last known value forecast')
plt.plot(test['Date'], naive_seasonal, label='Naive seasonal forecast')
plt.plot(test['Date'], historical_mean, label='Historical mean forecast')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock: AAPL - Daily Closing Price Forecast')
plt.xticks([], [])
plt.legend()
plt.show()
```

```python
#Forecast the next timestep over the test set, and measure the MSE. Did it decrease?
train_shift = train['Close'].shift(periods=1)
mse_one_step = mean_squared_error(test['Close'], train_shift[-5:])
print("The next timestep error: ", mse_one_step) 
```

Seems like it didn't decrease at all

```python
#Plot your forecasts
fig = plt.figure(figsize=(20, 10))
test['forecast'] = train_shift[-5:].values
plt.plot(train['Date'], train['Close'], label='Train Data')
plt.plot(test['Date'], test['Close'], label='Test Data')
plt.plot(test['Date'], test['forecast'], label='Forecast')
plt.xlabel('Date')
plt.ylabel('Closing price')
plt.title('One-Step Forecast')
plt.xticks([], [])
plt.legend()
plt.show()
```
